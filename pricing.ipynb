```python
df = df_original[(df_original.SegmentoEstrategico=='Branco') | (df_original.SegmentoEstrategico=='Amarelo')].groupby(['CdGrupoProdutoSimilar','NmMicroRegiao', 'NmMesoRegiao', 'NmProduto','IdProduto','IdCliente','NrPedido'])[['QtdeVendida','VlTotalBruto','VlReceitaLiquida']].sum()
df = df.reset_index()
df
```

```python
#df_sel = df[df['NmProduto']=='CARGA BOZZANO MONANG.BEST3 2X1']
df_sel = df[df['CdGrupoProdutoSimilar']==172]
#df_sel = df[df['IdProduto']==99353]
df_sel = df_sel[df_sel.VlTotalBruto!=0]
df_sel['PB'] = df_sel.VlTotalBruto/df_sel.QtdeVendida
df_sel
```

```python
pd.DataFrame(df_sel.groupby(['IdCliente'])['NrPedido'].count()).value_counts()


```

```python
df_aux = pd.DataFrame(df_sel.groupby(['IdCliente'])['NrPedido'].nunique())
df_aux = df_aux[(df_aux.NrPedido > 1)]
df_aux = df_aux.reset_index()
df_aux
```

```python
df_ana = df_sel[df_sel.IdCliente.isin(df_aux.IdCliente)]
df_ana
```

```python
import numpy as np
from tqdm import tqdm
from sklearn.linear_model import LinearRegression
model = LinearRegression()
intercepto = {}
coef = {}
total = {}
qutde = {}
meso = {}
micro = {}
segmento = {}
npedidos = {}
nome_produto = {}

for i in tqdm (range(0, 2123),desc="Processando Fornecedor", ascii=False, ncols=75):

   #df_aux = df[(df.CdGrupoProdutoSimilar==df_aux.CdGrupoProdutoSimilar[0])].groupby(['IdCliente'])[['NrPedido']].count()
   #df_aux_2 = df_aux[(df_aux.NrPedido)>1].reset_index()
   #df_aux_2

# Selecionando clientes de um mesmo grupo similar 8
   df_a = df_ana[(df_ana.IdCliente==df_ana.IdCliente.unique()[i])]

   x = np.array(np.log10(df_a.PB)).reshape(-1,1)
   y = np.array(np.log10(df_a.QtdeVendida)).reshape(-1,1)

   model.fit(x, y)

   sentence = [str(df_ana.IdCliente.unique()[i])]
   sentence = '-'.join(sentence)

   intercepto[sentence] = list(model.intercept_)
   coef[sentence] = list(model.coef_)[0][0]
   total[sentence] = df_a.VlTotalBruto.sum()
   qutde[sentence] = df_a.QtdeVendida.sum()
   meso[sentence] = list(df_a.NmMesoRegiao)[0]
   micro[sentence] = list(df_a.NmMicroRegiao)[0]
   segmento[sentence] = df_original[df_original.IdCliente==list(df_a.IdCliente)[0]][['NmMesoRegiao','NmMicroRegiao','SegmentoEstrategico']].iloc[0][2]
   nome_produto[sentence] = list(df_a.NmProduto)[0]
   npedidos[sentence] = df_a.NrPedido.count()
    
   #print(df_a.VlTotalBruto.sum())

```

```python
df_a
```

```python
final = pd.concat([pd.DataFrame.from_dict(coef,orient='index'),
           pd.DataFrame.from_dict(nome_produto,orient='index'),         
           pd.DataFrame.from_dict(total,orient='index'),
           pd.DataFrame.from_dict(qutde,orient='index'),
           pd.DataFrame.from_dict(meso,orient='index'),
           pd.DataFrame.from_dict(micro,orient='index'),
           pd.DataFrame.from_dict(segmento,orient='index'),
           pd.DataFrame.from_dict(npedidos,orient='index')
           ],axis=1)
final = final.reset_index()
final.columns = ['IdCliente','coef','nome_produto','total_valor','total_quantidade','meso_regiao','microregiao','segmento','num_de_pedidos']
```
